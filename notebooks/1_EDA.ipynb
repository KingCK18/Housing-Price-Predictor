```python
# Columns with too many missing values (drop them)
cols_to_drop = ["PoolQC", "MiscFeature", "Alley", "Fence"]

# Only drop columns that exist (safe for reruns)
cols_existing = [c for c in cols_to_drop if c in train.columns]

train = train.drop(columns=cols_existing)
# test = test.drop(columns=cols_existing)  # keep commented until test is loaded


# Fill categorical NA with "None"
cat_none = [
    "MasVnrType", "FireplaceQu",
    "GarageType", "GarageFinish", "GarageQual", "GarageCond",
    "BsmtQual", "BsmtCond", "BsmtExposure",
    "BsmtFinType1", "BsmtFinType2"
]

for col in cat_none:
    if col in train.columns:
        train[col] = train[col].fillna("None")
        # test[col] = test[col].fillna("None")


# Fill numeric NA values
if "MasVnrArea" in train.columns:
    train["MasVnrArea"] = train["MasVnrArea"].fillna(0)

if "LotFrontage" in train.columns:
    train["LotFrontage"] = train["LotFrontage"].fillna(train["LotFrontage"].median())


# Electrical: fill with mode
if "Electrical" in train.columns:
    train["Electrical"] = train["Electrical"].fillna(train["Electrical"].mode()[0])


# Drop Id column only if it exists
if "Id" in train.columns:
    train = train.drop(columns=["Id"])

```
